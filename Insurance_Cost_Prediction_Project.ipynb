```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
ins = pd.DataFrame(scaling.fit_transform(ins),columns=['age','sex','bmi','smoker','region','charges'],index=ins.index)
ins
```

```python
sn.distplot(ins.charges,color = "darkblue")
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set,test_set= train_test_split(ins,test_size=0.1)
```

```python
train_set_x=train_set.drop(["charges"],axis=1)
train_set_y=train_set["charges"]

test_set_x = test_set.drop(["charges"],axis=1)
test_set_y = test_set["charges"]  
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(train_set_x, train_set_y)
```

```python
some_data = train_set_x.iloc[:5]
some_labels = train_set_y.iloc[:5]
print("Predictions:", lin_reg.predict(some_data))
print("Labels:", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
ins_predictions = lin_reg.predict(train_set_x)
lin_mse = mean_squared_error(train_set_y, ins_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```
